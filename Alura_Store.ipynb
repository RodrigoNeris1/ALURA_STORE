```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
url1 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_1.csv'
url2 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_2.csv'
url3 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_3.csv'
url4 = 'https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/main/base-de-dados-challenge-1/loja_4.csv'

# Juntando todos os DataFrame dentro de um so
df = pd.concat([pd.read_csv(url1), pd.read_csv(url2), pd.read_csv(url3), pd.read_csv(url4)])

loja1 = pd.read_csv(url1)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)
```

```python
# FATURAMENTO DE TODAS AS LOJAS / ANALISE DE FATURAMENTO
lojas = {
    'Loja 1': loja1['Preço'].sum(),
    'Loja 2': loja2['Preço'].sum(),
    'Loja 3': loja3['Preço'].sum(),
    'Loja 4': loja4['Preço'].sum()
}

lojas
```

```python
# ANALISE DE FATURAMENTO POR PRODUTO E ESTADO

faturamento_por_produto = df.groupby(['Produto','Categoria do Produto'])['Preço'].sum().sort_values(ascending=False).head(10)

faturamento_por_produto
```

```python
# VENDAS POR CATEGORIA

vendas_categoria = df.groupby("Categoria do Produto")["Produto"].count().sort_values(ascending=False)

print(vendas_categoria)
```

```python
# CATEGORIAS MAIS VENDIDAS EM TODAS AS LOJAS (valor total)

categoria_mais_vendida = df.groupby(['Categoria do Produto'])['Preço'].sum().sort_values(ascending=False)

print(categoria_mais_vendida)
```

```python
# AVALIAÇAO DOS CLIENTES POR LOJA

# Adiciona coluna identificando a loja
loja1["loja"] = "loja1"
loja2["loja"] = "loja2"
loja3["loja"] = "loja3"
loja4["loja"] = "loja4"

# Concatena todos os DataFrames em um só
df = pd.concat([loja1, loja2, loja3, loja4], ignore_index=True)


avaliacao_media = df.groupby("loja")["Avaliação da compra"].mean().round(2)
print("\nMédia de avaliação das lojas:")

print(avaliacao_media)
```

```python
# GRAFICO - Média de Avaliação

medias = {
    'Loja 1': loja1['Avaliação da compra'].mean(),
    'Loja 2': loja2['Avaliação da compra'].mean(),
    'Loja 3': loja3['Avaliação da compra'].mean(),
    'Loja 4': loja4['Avaliação da compra'].mean()
}

# Cores para as fatias
cores = ['#2A73D9', '#4D4D4D', '#B0B0B0', '#8AB4F8']

# Calcular a porcentagem de cada média em relação ao total
total = sum(medias.values())
porcentagens = {loja: (media / total) * 100 for loja, media in medias.items()}

# Criando o gráfico de rosca
fig, ax = plt.subplots(figsize=(8, 6))

# Pie plot com largura reduzida para virar rosca
wedges, texts, autotexts = ax.pie(
    medias.values(),
    labels=medias.keys(),
    autopct='%1.1f%%',
    startangle=90,
    colors=cores,
    pctdistance=0.85,
    wedgeprops=dict(width=0.4, edgecolor='#B0B0B0')
)

# Adicionando círculo branco no centro para formar a rosca
centro = plt.Circle((0, 0), 0.60, fc='white')
fig.gca().add_artist(centro)
```

```python
# PRODUTOS MAIS E MENOS VENDIDO

produtos_vendidos = df["Produto"].value_counts()
mais_vendido = produtos_vendidos.idxmax()
menos_vendido = produtos_vendidos.idxmin()

print(f"\nPRODUTO MAIS VENDIDO: {mais_vendido} ({produtos_vendidos.max()} vendas)")
print(f"PRODUTO MENOS VENDIDO: {menos_vendido} ({produtos_vendidos.min()} vendas)")
```

```python
                                                                                # PLOTANDO OS PRODUTOS MAIS VENDIDOS POR LOJA
```

```python
# PRODUTOS MAIS VENDIDOS DA LOJA 1
produtos_sorted = loja1[['Produto', 'Preço']].groupby('Produto').sum().sort_values(by='Preço',ascending=False).head(10)
produtos_sorted
```

```python
# PLOTAGEM DO GRAFICO LOJA 1

fig, ax = plt.subplots(figsize=(12, 6))
ax.barh(produtos_sorted.index, produtos_sorted['Preço'])

# Título e rótulos dos eixos
ax.set_title('TOP 10 PRODUTOS MAIS VENDIDOS(LOJA 1)', loc='left', fontsize=18)
ax.set_xlabel('Total de Vendas (R$)', fontsize=14)
ax.set_ylabel('')
ax.invert_yaxis()
ax.xaxis.set_tick_params(labelsize=12)
ax.yaxis.set_tick_params(labelsize=12)

# Mostrar os valores ao lado das barras
for i, v in enumerate(produtos_sorted['Preço']):
  ax.text(v + 10, i, f'R$ {v:.2f}', va='center', fontsize=10)

# Melhorar a estética do gráfico
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)
ax.tick_params(axis='both', which='both', length=0)

plt.tight_layout()
plt.show()
```

```python
# GRAFICO DE EMPILHAMENTO

# Agrupando os dados: somando o faturamento por loja e por categoria
faturamento = df.groupby(['loja', 'Categoria do Produto'])['Preço'].sum().unstack().fillna(0)

# Paletta de cores (hexadecimais)
cores = ['#1f77b4', '#ff7f0e', '#2ca02c', '#ff4500', '#9467bd', '#b22222', '#e377c2', '#7f7f7f']

# Plotando gráfico de barras empilhadas
faturamento.plot(kind='bar', stacked=True, figsize=(12, 6), color=cores * len(faturamento.columns))

# Personalização do gráfico
plt.title('Faturamento por Loja e Categoria de Produto', fontsize=16)
plt.xlabel('Loja', fontsize=12)
plt.ylabel('Faturamento (R$)', fontsize=12 , labelpad=12)
plt.xticks(rotation=0)
plt.legend(title='Categoria do Produto', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```
